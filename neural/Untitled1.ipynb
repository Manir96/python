```python
# Load csv data
abalone_train = pd.read_csv('Sample-Dataset-Sheet4.csv',names=[
"BranchCode",
"LoanTerm",
"Gender",
"HomeType",
"FamilyMember",
"TotalWealth",
"PrincipalLoan",
"Frequency",
"PurposeCode",
"DFS_Status",
"Deposit	",
"Withdrawal",
"Balance",
"duedays",
"DueType",
"LoanStatus",
],)
```

```python
# Separate features and labels
abalone_features = abalone_train.copy()
abalone_labels = abalone_features.pop("LoanStatus")
```

```python
# Normalize features
abalone_features = ('abalone_features / abalone_features.max()')
```

```python
# Reshape features
abalone_features = np.array(abalone_features)
abalone_features = np.reshape(abalone_features, (-1, 28 * 28))
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
abalone_features, abalone_labels, test_size=0.2, random_state=42
)
```

```python
# Create ANN model
model = tf.keras.Sequential()
model.add(layers.Dense(128, activation="relu", input_shape=(784,)))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(64, activation="relu"))
model.add(layers.Dense(1))
```

```python
# Compile model
model.compile(optimizer="adam", loss="mse", metrics=["mae"])
```

```python
# Train model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1)
```

```python
# Evaluate model
model.evaluate(X_test, y_test)
```

```python
# Generate predictions
predictions = model.predict(X_test)
```
